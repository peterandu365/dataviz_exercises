```python
# import pandas
import pandas as pd

# import seaborn
import seaborn as sns

# import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('pokemon.csv',sep=',')
df.drop('#',axis=1,inplace=True)
```

```python
df.head()
```

### Task 1: Plot a regression line between attack and defense.

```python
# plot regression line for attack vs defense using seaborn
sns.lmplot(x='Attack',y='Defense',data=df)
```

### Task 2: Plot a regression line between attack and defense while displaying a distinction between the different generations. 

```python
# plot regression line for attack vs defense using seaborn, hue=Generation
sns.lmplot(x='Attack',y='Defense',data=df,hue='Generation')
```

### Task 3: Make a boxplot for each feature of dataframe (without Generation and Legendary features) in one plot.

```python
# plot boxplot for each column, except for generation and legendary:
sns.boxplot(data=df.drop(['Total', 'Generation','Legendary'],axis=1))
```

### Task 4: Plot a correlation heatmap of the pokemon dataframe (not including the features Generation and Legendary).

```python
# Plot a correlation heatmap of the pokemon dataframe, drop the generation and legendary columns:
sns.heatmap(df.drop(['Generation','Legendary'],axis=1).corr())
```
