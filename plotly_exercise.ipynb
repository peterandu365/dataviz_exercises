### Task 1: Create a scatterplot with a 1000 random data points. 

- x-axis values should come from a normal distribution using np.random.randn(1000)

- y-axis values should come from a uniform distribution over [0,1) using np.random.rand(1000)

```python
# obtain the x and y values:

# x is normal distribution using np.random.randn(1000)
# x = np.random.randn(1000)
x = np.random.randn(1000)   


# y is uniform distribution over [0,1] 
# y = np.random.rand(1000)
y = np.linspace(0, 1, 1000)
```

```python
# define a data variable
data = [go.Scatter(x=x, y=y, mode='markers')]
```

```python
# define the layout, include a title and axis labels
layout = go.Layout(title='Scatter Plot', xaxis=dict(title='Normal Distribution'), yaxis=dict(title='Uniform Distribution')) 
```

```python
# Create a fig from the data and layout. Then plot the fig
fig = go.Figure(data=data, layout=layout)
fig.show()
```

### Task 2: Create a stacked bar chart with the data from the link below. 

**Note** that the questions appear in the index and should be used for the x-axis, while the responses appear as column labels. You will need to create more `traces`, one for each column.

Stretch Task: Create a horizontal bar chart! Use this documentation [here](https://plot.ly/python/horizontal-bar-charts/) for help.

```python
# load data
import pandas as pd
df = pd.read_csv('mocksurvey.csv',index_col=0)
```

```python
df.columns.to_list()
df.head(1000)
```

```python
# create traces using a list comprehension:
# x = df's index values
# y = df's 1st colum values
# trace1 = "Stongly agree"

# ['Strongly Agree',
#  'Somewhat Agree',
#  'Neutral',
#  'Somewhat Disagree',
#  'Strongly Disagree']


trace1 = go.Bar(x=df.index, y=df['Strongly Agree'], name='Strongly Agree')
trace2 = go.Bar(x=df.index, y=df['Somewhat Agree'], name='Somewhat Agree')
trace3 = go.Bar(x=df.index, y=df['Neutral'], name='Neutral')
trace4 = go.Bar(x=df.index, y=df['Somewhat Disagree'], name='Somewhat Disagree')
trace5 = go.Bar(x=df.index, y=df['Strongly Disagree'], name='Strongly Disagree')

data = [trace1, trace2, trace3, trace4, trace5] 

```

```python

# trace1 = go.Bar(x=df.columns, y=df.iloc[0].values)
# trace2 = go.Bar(x=df.columns, y=df.iloc[1].values)
# trace3 = go.Bar(x=df.columns, y=df.iloc[2].values)
```

```python
# create a layout, remember to set the barmode here
layout = go.Layout(title='Mock Survey', barmode='stack')
```

```python
# create a fig from data & layout, and plot the fig.
fig = go.Figure(data=data, layout=layout)
fig.show()
```

### Task 3: Create a histogram that plots the `length` field from dataset below. Set the range from 0 to 1, with a bin size of 0.02.

```python
# load data
df = pd.read_csv('https://raw.githubusercontent.com/Pierian-Data/Plotly-Dashboards-with-Dash/master/Data/abalone.csv')

pd.DataFrame(df.columns)
```

```python
# create a data variable: Range: 0-1, bin size = 0.05
data = [go.Histogram(x=df['length'], xbins=dict(start=0, end=1, size=0.05))]
# data = go.Histogram(x=df['length'])
```

```python
# add a layout
Layout = go.Layout(title='Histogram of Length')
```

```python
# create a fig from data & layout, and plot the fig
fig = go.Figure(data=data, layout=Layout)
fig.show()
```
